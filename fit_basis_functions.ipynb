# Example of fitting basis functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Make some fake data

```python
x = np.arange(30)/30

# Linear term plus quadratic term plus gaussian noise
lnc = -3
qdc = 4
y = lnc*x + qdc*(x**2) + np.random.normal(scale=.1, size=len(x))

plt.plot(x,y, 'k.')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Noisy data to fit w/ basis functions")
plt.show()
```

## Make basis functions to fit data with

```python
b0 = x
b1 = x**2

plt.plot(x, b0, "k", label="Linear basis")
plt.plot(x, b1, "r", label="Quadratic basis")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
```

## Find best coefficents for basis functions

```python
G = np.vstack((b0, b1)).T
d = y

m = np.linalg.lstsq(G, d, rcond=None)[0]
print("Linear coefficent estimate: %.3f" % m[0])
print("Quadratic coefficent estimate: %.3f" % m[1])
# Similar to 
```

## Plot fit to data

```python
plt.plot(x,y, 'k.', label="data")
plt.plot(x, G@m, "r-", label="model")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Data and Model")
plt.legend()
plt.show()
```
