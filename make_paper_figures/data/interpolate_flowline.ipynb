```python
import pandas as pd
import numpy as np

# ---- Load CSV ----
df = pd.read_csv("flowline_5.csv")

# Extract x and y
x = df["x"].values
y = df["y"].values

# ---- Compute cumulative distance along path ----
dx = np.diff(x)
dy = np.diff(y)
dist = np.sqrt(dx**2 + dy**2)
cumdist = np.insert(np.cumsum(dist), 0, 0)

# ---- Create new denser distance grid ----
spacing = 50  # desired spacing in same units as x/y (e.g., meters)
new_dist = np.arange(0, cumdist[-1], spacing)

# ---- Interpolate x and y separately ----
x_new = np.interp(new_dist, cumdist, x)
y_new = np.interp(new_dist, cumdist, y)

# ---- Save to new CSV ----
df_new = pd.DataFrame({"x": x_new, "y": y_new})
df_new.to_csv("flowline_5_dense.csv", index=False)

print("Saved dense CSV as output_dense.csv")
```
