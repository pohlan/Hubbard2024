```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.backends.backend_pdf import PdfPages

# Create a figure
fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('off')

# Equation in center
ax.text(0.5, 0.5, r'$\frac{dR}{dt} = f - \nu$', fontsize=20, ha='center', va='center', color='navy')

# Draw a blue oval around the equation
ellipse = patches.Ellipse((0.5, 0.5), 0.5, 0.2, edgecolor='blue', facecolor='none', lw=2)
ax.add_patch(ellipse)

# Green text (terminus-related)
ax.text(0.2, 0.72, 'terminus\nposition', color='green', ha='center', va='center', fontsize=12)
ax.text(0.33, 0.56, 'terminus\nretreat rate', color='purple', ha='center', va='center', fontsize=12)

# Purple text (ablation and velocity)
ax.text(0.67, 0.56, 'frontal\nablation', color='purple', ha='center', va='center', fontsize=12)
ax.text(0.8, 0.52, 'ice velocity\n(supply)', color='purple', ha='center', va='center', fontsize=12)

# Bottom green text
ax.text(0.45, 0.32, 'melting', color='green', ha='center', va='center', fontsize=12)
ax.text(0.55, 0.32, 'calving', color='green', ha='center', va='center', fontsize=12)

# Gray text (environmental factors)
ax.text(0.25, 0.22, 'ocean temp', color='gray', ha='center', va='center', fontsize=11)
ax.text(0.2, 0.18, 'runoff', color='gray', ha='center', va='center', fontsize=11)
ax.text(0.14, 0.14, 'surface melt', color='gray', ha='center', va='center', fontsize=11)
ax.text(0.26, 0.14, 'liquid precip', color='gray', ha='center', va='center', fontsize=11)

# Orange/red labels (buttressing, basal motion)
ax.text(0.8, 0.7, 'buttressing', color='orangered', ha='center', va='center', fontsize=12)
ax.text(0.8, 0.35, 'basal\nmotion', color='orangered', ha='center', va='center', fontsize=12)

# Arrows for logical connections
arrowprops = dict(arrowstyle="->", lw=1.5, color='green')
ax.annotate('', xy=(0.5, 0.6), xytext=(0.2, 0.68), arrowprops=arrowprops)
ax.annotate('', xy=(0.35, 0.5), xytext=(0.33, 0.54), arrowprops=arrowprops)
ax.annotate('', xy=(0.65, 0.5), xytext=(0.67, 0.54), arrowprops=arrowprops)
ax.annotate('', xy=(0.5, 0.4), xytext=(0.45, 0.34), arrowprops=arrowprops)
ax.annotate('', xy=(0.5, 0.4), xytext=(0.55, 0.34), arrowprops=arrowprops)

# Save to PDF
#pdf_path = "glacier_terminus_dynamics_diagram.pdf"
#with PdfPages(pdf_path) as pdf:
#    pdf.savefig(fig, bbox_inches='tight')

#plt.close(fig)
#print(f"Saved to {pdf_path}")
```
